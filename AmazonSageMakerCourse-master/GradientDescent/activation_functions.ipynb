<h4>Activation Functions</h4>
<ul>
<li>Introduces non-linearity in a neural network</li>

<li>Improves a models to fit complex non-linear datasets</li>

<li>Three popular activation functions: sigmoid, relu, tanh</li>
</ul>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sigmoid or logistic function
# For any input, output is bounded between 0 & 1
# Output is:
#  < 0.5 for negative values
#  0.5 when input is 0
#  > 0.5 for positive values

def sigmoid_func(x):
    return 1.0/(1 + np.exp(-x))
```

```python
# rectified linear unit (ReLU)
# For any input x, output is max(0, x)
# Output is:
#  0 for negative values
#  x for any non-negative x

def relu_func(x):
    return max(x,0)
```

```python
# tanh
# For any input x, output is bounded between -1 and 1
def tanh_func(x):
    return np.tanh(x)
```

```python
numbers = [10,0,-10]

print('Number\tSigmoid\tRelu\tTanh')
for i, n in enumerate(numbers):
    print('{0}\t{1:5.5f}\t{2:5.2f}\t{3:5.5f}'.format(n, 
                                           sigmoid_func(n),
                                           relu_func(n),
                                           tanh_func(n)
                                          ))
    
```

```python
# Sigmoid function example
x = pd.Series(np.arange(-8,8,0.5))
y_sigmoid = x.map(sigmoid_func)
y_relu = x.map(relu_func)
y_tanh = x.map(tanh_func)
```

```python
plt.plot(x,y_sigmoid,label='sigmoid')
plt.plot(x,y_tanh,label='tanh')

plt.xlabel('input')
plt.ylabel('output')
plt.axhline(y=0,ls='dashed',alpha=0.6)
plt.axvline(x=0,ls='dashed',alpha=0.6)
plt.grid(True)

plt.title('Activation Functions')
plt.legend()
plt.show()
```

```python
plt.plot(x,y_relu,label='relu')

plt.xlabel('input')
plt.ylabel('output')
plt.axhline(y=0,ls='dashed',alpha=0.6)
plt.axvline(x=0,ls='dashed',alpha=0.6)
plt.grid(True)

plt.title('Activation Functions')
plt.legend()
plt.show()
```
